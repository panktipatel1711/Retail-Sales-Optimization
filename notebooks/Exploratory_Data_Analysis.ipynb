# Exploratory Data Analysis (EDA) - Retail Sales Forecasting
This notebook contains the initial analysis of the retail sales data to understand trends, seasonality, and prepare the data for the Random Forest model.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the generated dataset
df = pd.read_csv('../data/raw_sales_data.csv')
df['Date'] = pd.to_datetime(df['Date'])

print("Dataset Overview:")
print(df.head())
print("\nSummary Statistics:")
print(df.describe())
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Date', y='Sales', color='blue')
plt.title('Daily Sales Trend (Historical Data)')
plt.xlabel('Date')
plt.ylabel('Units Sold')
plt.grid(True)
plt.show()
```

```python
# Testing lag features for the ML model
df['Day_of_Week'] = df['Date'].dt.dayofweek
df['Lag_1'] = df['Sales'].shift(1)

plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x='Day_of_Week', y='Sales')
plt.title('Sales Distribution by Day of Week')
plt.show()
```
